```python
# Exercise 1: Read a file and display its contents

def read_file(filename):
    try:
        with open(filename, 'r') as file:
            content = file.read()
            print(content)
    except FileNotFoundError:
        print("File not found.")

filename = filename = r"C:\Users\pssha\OneDrive\Desktop\covid_19_india.csv"

read_file(filename)
```

```python
# Exercise 2: Copy contents of one file to another

def copy_file(source, destination):
    try:
        with open(source, 'r') as src, open(destination, 'w') as dest:
            dest.write(src.read())
        print("File copied successfully.")
    except FileNotFoundError:
        print("Source file not found.")

source_file = r"C:\Users\pssha\OneDrive\Desktop\covid_19_india.csv"
destination_file = r"C:\Users\pssha\OneDrive\Desktop\work 8 bit\empty.csv"
copy_file(source_file, destination_file)
```

```python
# Exercise 3: Count the total number of words in a file

def count_words(filename):
    try:
        with open(filename, 'r') as file:
            content = file.read()
            words = content.split()
            print("Total number of words:", len(words))
    except FileNotFoundError:
        print("File not found.")

filename = r"C:\Users\pssha\OneDrive\Desktop\work 8 bit\count.txt"
count_words(filename)
```

```python
def convert_to_integer():
    try:
        user_input = input("Enter a number: ").strip()  # Remove extra spaces
        
        # Remove surrounding quotes if present
        if (user_input.startswith(("'", '"')) and user_input.endswith(("'", '"'))):
            user_input = user_input[1:-1]
        
        num = int(user_input)  # Convert to integer
        print("Converted integer:", num)
    
    except ValueError:
        print("Invalid input! Please enter a valid integer.")

convert_to_integer()
```

```python
# Exercise 5: Raise an exception if the list contains negative numbers

def check_positive_numbers():
    try:
        numbers = list(map(int, input("Enter a list of integers (space-separated): ").split()))
        if any(num < 0 for num in numbers):
            raise ValueError("Negative numbers are not allowed.")
        print("All numbers are positive.")
    except ValueError as e:
        print("Error:", e)

check_positive_numbers()
```

```python
# Exercise 6: Compute the average with exception handling

def compute_average():
    try:
        numbers = list(map(int, input("Enter a list of integers (space-separated): ").split()))
        avg = sum(numbers) / len(numbers)
        print("Average:", avg)
    except ZeroDivisionError:
        print("Error: Cannot divide by zero (list is empty).")
    except ValueError:
        print("Error: Invalid input. Please enter integers only.")
    finally:
        print("Program execution completed.")

compute_average()
```

```python
# Exercise 7: Write a string to a file with exception handling

def write_to_file():
    try:
        filename = "D:/DATA SCIENCE/PRACTISE/ADDRESS.txt"
        content = input("Enter text to write into the file: ")
        with open(filename, 'w') as file:
            file.write(content)
        print("File written successfully. Welcome!")
    except Exception as e:
        print("An error occurred:", e)

write_to_file()
```
